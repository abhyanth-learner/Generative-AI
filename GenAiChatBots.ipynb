```python
import os 
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_groq import ChatGroq
```

```python
os.environ["GROQ_API_KEY"]="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"
```

```python
llm=ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0.1,
)

parser=StrOutputParser()
```

```python
def chat():
    chatHistory=[
        ("system","You ar a helpful chatbot.Give answer only related to cricket any other topic please respond with Sorry and some message of you own.")
    ]

    print("ChatBot:Type Ext to quit\n")

    while True:
        user_input=input("You: ").strip()

        if user_input.lower() =="exit":
            break

        chatHistory.append(("user",user_input))

        prompt=ChatPromptTemplate.from_messages(chatHistory)

        chain=prompt | llm | parser

        response=chain.invoke({})

        print(f"Bot:{response}\n")

        chatHistory.append(("assistant",response))

        print("-"*100)
```

```python
chat()
```

```python
import os 
from llama_index.core.llms import ChatMessage,MessageRole
from llama_index.llms.groq import Groq
```

```python
os.environ["GROQ_API_KEY"]="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"
llm=Groq(
    model="llama-3.3-70b-versatile",
    temperature=0.1,
)
```

```python
def chat():
    history=[
        ChatMessage(role=MessageRole.SYSTEM,content="You ar a helpful chatbot.Give answer only related to cricket any other topic please respond with Sorry and some message of you own.")
    ]

    print("Llama Chatbot. Type 'exit' to quit\n")

    while True:
        user_input=input("You: ").strip()
        if user_input.lower() in {"exit","quit"}:
            break

        history.append(ChatMessage(role=MessageRole.USER,content=user_input))

        response=llm.chat(messages=history)
        answer=response.message.content

        print(f"Bot:{answer}\n")

        history.append(ChatMessage(role=MessageRole.ASSISTANT,content=answer))

        print("-"*80)
```

```python
chat()
```
